```python
import papermill as pm
import os

input_notebook = "aoi.ipynb"

os.makedirs('outputs', exist_ok=True)

for i in range(5):
    # 為每次執行指定不同的輸出文件名
    output_notebook = f"outputs/output_notebook.ipynb"
    
    # 執行 Notebook 並傳遞參數
    pm.execute_notebook(
        input_path=input_notebook,
        output_path=output_notebook,
        parameters=dict(param1=i)
    )
```
