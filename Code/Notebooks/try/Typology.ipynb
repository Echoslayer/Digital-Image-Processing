```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 讀取圖像
image = cv2.imread('F:\\Code\\Digtal-Image-Processing\\Code\\Resources\\Data\\LAB_CAT2.jpg', cv2.IMREAD_GRAYSCALE)

# 定義結構元素
kernel = np.ones((5, 5), np.uint8)

# 開放操作
opening = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)

# 封閉操作
closing = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)

# 使用 matplotlib 顯示結果
plt.figure(figsize=(10, 7))

plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Opening')
plt.imshow(opening, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Closing')
plt.imshow(closing, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
