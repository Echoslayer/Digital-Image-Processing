```python
import cv2  # OpenCV library for image processing
import numpy as np  # NumPy library for array manipulation
import random  # For generating random values
import matplotlib.pyplot as plt  # Matplotlib library for displaying images
```

```python
# 讀取圖片
image = cv2.imread('F:/Code/Digital-Image-Processing/Code/Resources/Data/man.jpg')

# # 將圖片轉為灰度圖像（選擇性，如果你想在灰度圖上添加雜訊）
# gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # 將圖像從 BGR 轉換為 RGB，適應 plt 顯示
```

```python
# 定義胡椒鹽雜訊的參數
salt_pepper_ratio = 1  # 胡椒和鹽的比例
amount = 0.1  # 雜訊佔圖像的比例

# 創建與原圖像相同大小的空白雜訊圖像
noisy_image = np.copy(image)
total_pixels = image.shape[0] * image.shape[1]  # 計算總像素數
num_salt = int(total_pixels * amount * salt_pepper_ratio)  # 計算鹽雜訊數量
num_pepper = int(total_pixels * amount * (1.0 - salt_pepper_ratio))  # 計算胡椒雜訊數量

# 添加鹽雜訊（白色點）
for _ in range(num_salt):
    i = random.randint(0, image.shape[0] - 1)
    j = random.randint(0, image.shape[1] - 1)
    noisy_image[i, j] = [255, 255, 255]  # 將像素值設為白色

# 添加胡椒雜訊（黑色點）
for _ in range(num_pepper):
    i = random.randint(0, image.shape[0] - 1)
    j = random.randint(0, image.shape[1] - 1)
    noisy_image[i, j] = [0, 0, 0]  # 將像素值設為黑色

# 使用 plt 顯示圖片
plt.figure(figsize=(10, 6))
plt.imshow(noisy_image)
plt.title('Image with Salt and Pepper Noise')
plt.axis('off')  # 關閉坐標軸

plt.savefig(r"F:\Code\Digital-Image-Processing\Code\Resources\Data\noisy_man.jpg", bbox_inches='tight')

plt.show()
```
