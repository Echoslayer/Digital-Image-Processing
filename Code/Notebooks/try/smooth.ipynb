```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 讀取圖像並轉換為灰度圖像
img = cv2.imread('F:\Code\Digtal-Image-Processing\Code\Resources\Data\LAB_CAT2.jpg', cv2.IMREAD_GRAYSCALE)

# 計算原始圖像的標準差
original_std = np.std(img)

# 應用 3x3 平滑濾波器
smoothed_img = cv2.blur(img, (3, 3))

# 計算平滑後圖像的標準差
smoothed_std = np.std(smoothed_img)

# 計算標準差變化比例
std_ratio = smoothed_std / original_std

# 顯示原始圖像和平滑後圖像
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title(f'原始圖像\n標準差: {original_std:.2f}')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(smoothed_img, cmap='gray')
plt.title(f'平滑後圖像\n標準差: {smoothed_std:.2f}')
plt.axis('off')

plt.show()

# 打印標準差變化比例
print(f'平滑後標準差是原標準差的比例: {std_ratio:.2f}')
```
