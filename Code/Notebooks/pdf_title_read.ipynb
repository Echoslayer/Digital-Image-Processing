```python
import fitz  # PyMuPDF 的導入名稱

def extract_titles_from_pdf(pdf_path):
    # 打開 PDF 文件
    pdf_document = fitz.open(pdf_path)
    
    # 提取第一頁的文本
    page = pdf_document.load_page(0)
    text = page.get_text()

    # 分析文本並提取標題
    titles = []
    lines = text.split('\n')
    
    for line in lines:
        if line.startswith("Part"):
            titles.append(line)
    
    return titles

# 使用示例
pdf_path = r"F:\Code\Digital-Image-Processing\content\Books\Back to Basics\PDF\01.pdf"  # 替換為你的 PDF 文件路徑
titles = extract_titles_from_pdf(pdf_path)

# 輸出標題
for title in titles:
    print(title)
```

```python
import os

dir_path = r"F:\Code\Digital-Image-Processing\content\Books\Back to Basics\PDF"
for name in os.listdir(dir_path):
    if name.endswith('.pdf'):
        pdf_path = os.path.join(dir_path, name)
        
    titles = extract_titles_from_pdf(pdf_path)

    # 輸出標題
    for title in titles:
        print(title)
```

```python
import os

def rename_files_in_folder(folder_path):
    # 確保指定的資料夾存在
    if not os.path.isdir(folder_path):
        print(f"資料夾 {folder_path} 不存在。")
        return
    
    # 遍歷資料夾中的所有檔案
    for filename in os.listdir(folder_path):
        # 構建舊檔案的完整路徑
        old_file_path = os.path.join(folder_path, filename)
        
        # 確保這是一個檔案（而不是子資料夾）
        if os.path.isfile(old_file_path):
            # 分離檔案名稱和副檔名
            name, ext = os.path.splitext(filename)
            
            # 構建新檔案的名稱
            new_filename = f"{name}_b2bsk{ext}"
            
            # 構建新檔案的完整路徑
            new_file_path = os.path.join(folder_path, new_filename)
            
            # 重命名檔案
            os.rename(old_file_path, new_file_path)
            print(f"檔案 {filename} 已重命名為 {new_filename}")

# 使用示例
dir_path = r"F:\Code\Digital-Image-Processing\content\Books\Back to Basics\PDF"  # 替換為你的資料夾路徑
rename_files_in_folder(dir_path)
```
